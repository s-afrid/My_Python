### String methods

### Math functions
The math functions are located in math module, we have to import using `import`

````python
import math
# Round a floating point number to nearest whole integer
print(round(3.14))
# Round up using ceil
print(math.ceil(3.14))
# Round down using floor
print(math.floor(3.14))
# Give absolute value of a number
print(abs(-3.14))
# Raise a base number to a power
print(pow(3,2))
# Return square root of a number
print(math.sqrt(49))
# Return maximum value from a group of values
print(max(1,2,4))
# Return minimum value from a group of values
print(min(1,2,3))
````

### String slicing
Slicing is create a substring by extracting elements from another string using indexing or slice() method



````python
# Using indexing
# [start:stop:step]
name = "Bro Code"
first = name[0:3] # stopping index should be incremented by 1
first = name[:3] # excluding first indexing then python assumes it as 0
last = name[4:8]
last = name[4:] # excluding last indexing then python assumes it as n+1 (till the end)
print(first,last)

# Step is an optional field which is how much we are increasing our index by between starting and stopping, by default its 1.
funcky = name[0:8:2]
funcky = name[::2] # assume starting and last
print(funcky)

# To reverse the string
rev = name[::-1]
print(rev)

# Using slice method
website = "https://google.com"
sliceobj = slice(7,-4) #(start, stop)
print(website[sliceobj])
````

## Conditionals and Loops

### If statements
A block of code that will execute if it's condition is true, else block is conditionis false.

````python
age = 17
if age == 100:
    print("You are a century old!")
elif age >= 18:
    print("You are an adult")
elif age < 0:
    print("You haven't born yet!")
else:
    print("You are a child!")
````



Ta add second condition if 1st condition becomes false we use "else-if" or for short "elif"

Order of the statements matters!

***Logical Operators*** (and, or, not) used to check if two or more conditional statements are true.

````python
temp = 30
if temp >= 0 and temp <= 30:
    # Both the conditions must be true for overall condition to be true
    print("The temperature is good today!")
    print("Go outside")
elif temp < 0 or temp > 30:
    # Any one condition can be true then overall condition to be true
    print("Stay inside")
    print("The temperature is bad today")

# The not will flip to false if the condition is true
if not(temp>=0 and temp <= 30):
    print("stay inside")
````

### While loop
A statement that will execute it's block of code, as long as it's condition remains true.
```python
while 1==1:
    print("Help I'm stuck in a loop!")
```
This is an example of infinite loop where condition is always true.

````python
name = "" # Empty name variable
while len(name)==0:
    # Untill the user gives some value we are going keep asking him
    name = input("Enter your name: ")

# After while loop is escaped the name is printed using print
print("Hello "+name)
````

````python
name = None # None represents empty variable, it means false
while not name:
    name = input("Enter your name: ")
print("Hello "+name)
````

### For loop
A statement that will execute it's block of code a limited amount of time

while = unlimited  
for = limited

````python
for i in range(5):
    print(i)
````

i can be any variable, it is called a loop counter  
range(5) generates number from 0-4  

`range(50,100,2)`  
50 - first number is inclusive  
100 - last number exclusive, means if 100 then take 100-1  
2 - steps skip every 2nd number

````python
for i in range(1,6,2):
    print(i)
````

Using for loop we can iterate through anything that is considered iterable, this could be a string

````python
for i in "Bro":
    # i goes through each letter of the string and print's it
    print(i)
````

***Nested loops***, the "inner loop" will finish all of it's iterations before finishing one iteration of the "outer loop"

````python
# Lets create a rectangle with symbols
height = 5
width = 5
symbol = "#"
for i in range(height):
    # The outer for loop to take incharge of the height
    # Loop as the number of times as the height
    for j in range(width):
        # The inner loop to take incharge of width
        # Loop as the number of time as width
        print(symbol,end="")
    print()
````

`end=""` present going to next line for each iteration of inner loop  
`print()` empty arguments means print nothing and goes to new line after each outer iteration of outer loop

### Loop control statement
Change a loop execution from it's normal sequence.

- `break` : Used to terminate the loop entirely
- `continue` : Skips to the next iteration of the loop
- `pass` : Does nothing, act as a placeholder.

````python
while True:
    name = input("Enter your name: ")
    if name != "":
        # if name is not empty break is executed and while loop is terminated
        break
print(name)
````

````python
phone = "123-456"
for i in phone:
    if i == "-":
        # if i encounters '-' in phone number string , then skip current iteration and move on to next
        continue
    print(i,end=" ")
print()

for i in range(1,11):
    if i==5:
        # if i encounters 5 just do nothing
        pass
    else:
        print(i,end=" ")
````

### Set
Collection which is unordered, unindexed, no duplicate values

````python
utensiles = {"Fork","Spoon","Knife"}
# Set is created using flower braces
print(utensiles)
````

````python
utensiles.add("Napkin") # add new element to the set
print(utensiles)
utensiles.remove("Fork") # remove existing element from the set
print(utensiles)

# Consider a second set
dishes = {"Boul","Plate","Cup"}
utensiles.update(dishes) # add all element present in one set to other set
print(utensiles)

dinner_table = utensiles.union(dishes) # Make union of utensiles and dishes and store in dinner_table set
print(dinner_table)
print(utensiles.difference(dishes)) # What utensiles has but dishes doesn't. Change the position and vice versa can be checked
print(utensiles.intersection(dishes)) # What is the common element between two sets
````

### Dictionaries
A changable, unordered collection of unique key:value pairs. Fast because they use hashing, allow us to access a value quickly.





````python
capitals = {"USA":"Washington DC","India":"New Delhi","China":"Beijing","Russian":"Moscow"}
# Key - value separated by colon
# Key now become index of the value
print(capitals)
print(capitals["Russian"]) # While indexing, instead of index number we use the key
````

````python
print(capitals.get('Germany')) # Check if key 'Germany' exists in the dictionary or not
print(capitals.keys()) # Return all the keys that exist in a dictionary
print(capitals.values()) # Return all the values that exist in a dictionary
print(capitals.items()) # Return all the values that exists
````

````python
# Dictionary like other collection is also iterable
for key, value in capitals.items():
    print(key,"-", value)
````

````python
# Dictionaries are mutable like lists, they are changable

capitals.update({"Germany":"Berlin"}) # Add a new key, value pair to existing dictionary
print(capitals)
capitals.update({"USA":"Las Vegas"}) # Also update existing ones
print(capitals)

capitals.pop('China') # Remove key, value pair
print(capitals)
capitals.clear() # Clear whole dictionary
print(capitals)
````

## Functions
A block of code which is executed only when it is called. With functions we do not hane to execute a piece of code again and again.

````python
def hello():
    # def is used here to define the function and hello is the function name
    print("Hello")

# With only function name and paranthesis we call the function (invoke a function) and it run the code inside it
hello()
````

We can send our function same data or information and function can do something with that, the values we send to a funtion are called ***arguments***. At function definationation we need matching variables to accept that value, they are called ***parameters***.

````python
def hello(name):
    # name is parameter
    print("Hello "+name)

hello("Bro") # Bro is the argument
hello("Purple") # can send different argument

n1 = "King"
hello(n1) # can send variables too
````

A function can have multiple parameters and at the same time we send the same number of arguments

````python
def hello(first, last):
    print("Hello", first, last)
    # 2 parameters

# 2 arguments
hello("Bro","Code")
````

````python
# can send different type of data too
def hello(first, last, age):
    print("Hello",first,last)
    print("You are",age,"years old")

hello("Bro","Code",21)
````

### Return type
Function send python values or objects back to the caller. These values/objects are known as the function return values.


````python
def multiply(n1,n2):
    result = n1 * n2
    # The value result is returned to the caller
    return result

print(multiply(3,2)) # Print the value that is returned after function call

x = multiply(6,8) # store the return value
print(x)
````
