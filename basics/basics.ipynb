### String methods
Consider a string  
name = "Bro"

````python
name = "Bro"
# len() function will return the length of string
print(len(name))
# find() returns the index where the character is in the string
print(name.find("B"))
# Convert the first character of the string to capital
print(name.capitalize())
# Make all characters of the string as upper case
print(name.upper())
# Make all charaters of the string as lower case
print(name.lower())
# Check whether the variable is a digit
print(name.isdigit())
# Check whether the variable is a alphabet
print(name.isalpha()) # will return false if space or number or special characters are there
# Print the number of occurances of the character in string
print(name.count("o"))
# Replace all the occurances of the character with other character
print(name.replace("o","a"))
# Check whether the string is alphanumeric
print(name.isalnum())
# Print a string multiple times
print(3*name)
````

### Math functions
The math functions are located in math module, we have to import using `import`

````python
import math
# Round a floating point number to nearest whole integer
print(round(3.14))
# Round up using ceil
print(math.ceil(3.14))
# Round down using floor
print(math.floor(3.14))
# Give absolute value of a number
print(abs(-3.14))
# Raise a base number to a power
print(pow(3,2))
# Return square root of a number
print(math.sqrt(49))
# Return maximum value from a group of values
print(max(1,2,4))
# Return minimum value from a group of values
print(min(1,2,3))
````

### String slicing
Slicing is create a substring by extracting elements from another string using indexing or slice() method



````python
# Using indexing
# [start:stop:step]
name = "Bro Code"
first = name[0:3] # stopping index should be incremented by 1
first = name[:3] # excluding first indexing then python assumes it as 0
last = name[4:8]
last = name[4:] # excluding last indexing then python assumes it as n+1 (till the end)
print(first,last)

# Step is an optional field which is how much we are increasing our index by between starting and stopping, by default its 1.
funcky = name[0:8:2]
funcky = name[::2] # assume starting and last
print(funcky)

# To reverse the string
rev = name[::-1]
print(rev)

# Using slice method
website = "https://google.com"
sliceobj = slice(7,-4) #(start, stop)
print(website[sliceobj])
````

## Conditionals and Loops

### If statements
A block of code that will execute if it's condition is true, else block is conditionis false.

````python
age = 17
if age == 100:
    print("You are a century old!")
elif age >= 18:
    print("You are an adult")
elif age < 0:
    print("You haven't born yet!")
else:
    print("You are a child!")
````



Ta add second condition if 1st condition becomes false we use "else-if" or for short "elif"

Order of the statements matters!

***Logical Operators*** (and, or, not) used to check if two or more conditional statements are true.

````python
temp = 30
if temp >= 0 and temp <= 30:
    # Both the conditions must be true for overall condition to be true
    print("The temperature is good today!")
    print("Go outside")
elif temp < 0 or temp > 30:
    # Any one condition can be true then overall condition to be true
    print("Stay inside")
    print("The temperature is bad today")

# The not will flip to false if the condition is true
if not(temp>=0 and temp <= 30):
    print("stay inside")
````
